```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
s = np.loadtxt('biggrid.text', dtype=np.float, comments='!')
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(s[:, 0], s[:, 1], s[:, 2], s=20*s[:, 3]/np.linalg.norm(s[:, 3], np.inf))
#ax.scatter(2*s[:, 0], 2*s[:, 1], 2*s[:, 2], )
#ax.scatter(3*s[:, 0], 3*s[:, 1], 3*s[:, 2])
#ax.scatter(6*s[:, 0], 6*s[:, 1], 6*s[:, 2])
ax.scatter(0, 0, 0)
```
