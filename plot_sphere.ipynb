```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
s = np.loadtxt('biggrid.text', dtype=np.float, comments='!')
print("Loaded {} grid points.".format(len(s)))
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(s[:, 0], s[:, 1], s[:, 2], s=20*s[:, 3]/np.linalg.norm(s[:, 3], np.inf))
ax.scatter(0, 0, 0)
```

```python
try:
    for point in s:
        assert(np.linalg.norm(point[0:3]) - 1.0 < 1.0E-10)

    print('It really is a sphere! 👌')
except AssertionError:
    print('Oh noes! Something is wrong! 🐓')
```
