# Coulomb potential

```python
from scipy.optimize import minimize
from scipy.special import erf, erfc
def error(fac, r, pot):
    return np.sum( (pot-(fac/r))**2 )
```

```python
a = np.loadtxt('../coulomb')
r = a[:,0]
pot = a[:,1]

r_f = a[60:,0]
pot_f = a[60:,1]

res = minimize(error, 1.0, args=(r_f, pot_f))
print(res)

plt.figure(figsize=(12,5))

plt.plot(r, pot, '.-', label='Potential')
plt.plot(r_f, res.x*1./r_f, 'x-', label='Fit')
plt.plot(r, 2*np.pi*erf(r)/r, 'x-', label='$2\pi erf(r)/r$')

plt.xlim(0, 20)
plt.ylim(0, 7)

plt.axhline(np.pi*2, color='#333333', ls='-.', label='$2\pi$')

plt.grid()
plt.legend()
```
