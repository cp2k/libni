```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
s = np.loadtxt('twoatom_grid.text', dtype=np.float, comments='!')
print("Loaded {} grid points.".format(s.shape))
```

```python
len(np.unique(s, axis=0))-len(s)
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(s[:, 0], s[:, 1], s[:, 2],
           s=50*s[:, 3]/np.linalg.norm(s[:, 3], np.inf))
#ax.scatter(0, 0, 0)
```

```python
try:
    for point in s:
        assert(np.linalg.norm(point[0:3]) - 1.0 < 1.0E-10)

    print('It really is a sphere! 👌')
except AssertionError:
    print('Oh noes! Something is wrong! 🐓')
```

```python
fig = plt.figure(figsize=(12, 6))
n = 100
x_axis = np.arange(1, n)
plt.plot(x_axis, np.sin(np.pi*(2*x_axis-1)/(2*n)), '.')
```

```python
fig = plt.figure(figsize=(12, 6))
n = 100
x_axis = np.arange(1, n)
plt.plot(x_axis, np.cos(np.pi*(2*x_axis-1)/(2*n)), '.')
```

```python
def part(x):
    return 2*x**3 - 3*x**2 + 1

fig = plt.figure(figsize=(12, 6))
n = 100
x_axis = np.linspace(0, 1, n)
plt.plot(x_axis, part(part(part(x_axis))), '-.')
plt.plot(x_axis, part(x_axis), '-.')
plt.grid()
```
